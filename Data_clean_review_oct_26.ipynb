```python
## load the data into sentences
from nltk.tokenize import sent_tokenize

def LoadData(fpath):
    f=open(fpath)
    text=f.read().strip()
    f.close()
    sentences = nltk.sent_tokenize(text)
    return sentences
```

```python
## split the txt into sentence, sentence into word

from nltk.corpus import stopwords
from many_stop_words import get_stop_words
import re

def cleanData(sentences):
    results = []
    stopword = []
    stopword = list(stopwords.words('english'))## about 150
    nltk_words = list(get_stop_words('en'))#about 900
    stopword.extend(nltk_words)
#     print(stopword)
    for sentence in sentences:
        sentence=sentence.lower().strip()
        sentence=re.sub('[^a-z]',' ',sentence)#replace chars that are not letters or numbers with a space  
        words=sentence.split()
        for word in words:
            if word in stopword: continue
            else: results.append(word)                
    return results
```

```python
## get two combine word of the review

from nltk.util import ngrams

def GetTwoGrams(words):
    bigrams = ngrams(words,2)
    return bigrams
```

```python
## get the frequncy of the word

import collections

def CountFreq(words):
    counter=collections.Counter(words)
    return counter
```

```python
## plot word cloud
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

from subprocess import check_output
from wordcloud import WordCloud, STOPWORDS

def plotWordCloud(word):
    mpl.rcParams['figure.figsize']=(16.0,16.0)    #(6.0,4.0)
    mpl.rcParams['font.size']=12                #10 
    mpl.rcParams['savefig.dpi']=100             #72 
    mpl.rcParams['figure.subplot.bottom']=.1 
    stopwords = set(STOPWORDS)
    

    wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=200,
                          max_font_size=40, 
                          random_state=42
                         ).generate(word)

    print(wordcloud)
    fig = plt.figure(1)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
    fig.savefig("word1.png", dpi=900)
```
